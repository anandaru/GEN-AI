```python
from langchain.embeddings import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
data = pd.read_excel('/content/drive/MyDrive/words.xlsx')
print(data)
```

```python
data['embedding'] = data['Words'].apply(lambda x: embeddings.embed_query(x))
data['embedding']
```

```python
data.to_csv('/content/drive/MyDrive/embedding.csv')
```

```python
new_df= pd.read_csv('/content/drive/MyDrive/embedding.csv')
print(new_df)
```

```python
mytext = 'badminton'
text_embedding = embeddings.embed_query(mytext)
```

```python
print(f'the embedding is {text_embedding}')
```

```python
from scipy.spatial.distance import cosine
```

```python
data['similarity_score'] = data['embedding'].apply(lambda x: 1-cosine(text_embedding, x))
data
```

```python
data.sort_values('similarity_score',ascending=False)
```
